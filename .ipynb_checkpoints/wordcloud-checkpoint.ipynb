```python
import pandas as pd
import json
from pandas.io.json import json_normalize

with open('fleetx.ghost.2020-07-18-20-18-49.json', encoding='utf-8') as f: 
    d = json.load(f)  
nycphil = json_normalize(d['db'][0]['data']['posts']) 
nycphil.head()
```

```python
nycphil.plaintext=nycphil.plaintext.str.replace('\n', ' ').replace('\xa0',' ').replace('.', ' ')
nycphil.dropna(subset=['plaintext'], axis=0, inplace=True)
nycphil.plaintext
```

```python
nycphil.iloc[10]['plaintext']
```

```python
x=nycphil.plaintext.to_string(index=None)
```

```python
x=x.replace('\n', " ")
```
